```python
import sys
sys.path.append('/app')
from src.voice.stt import transcribe_audio
from pathlib import Path
import asyncio
```

```python
# CrÃ©er un audio de test
import edge_tts

text = "Bonjour, test de transcription"
audio_path = Path('/tmp/test.mp3')

async def create():
    tts = edge_tts.Communicate(text, 'fr-FR-DeniseNeural')
    await tts.save(str(audio_path))

await create()
print(f"Audio crÃ©Ã©: {audio_path}")
```

```python
# Transcrire
import time

start = time.time()
transcription = await transcribe_audio(audio_path, 'fr')
duration = time.time() - start

print(f"Original: {text}")
print(f"Transcription: {transcription}")
print(f"Temps: {duration:.2f}s")
```
